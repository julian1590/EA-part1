```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
import sys
sys.path.insert(0, 'EA1')
import plotly.express as px
import plotly.graph_objs as go
import json
import os
import pandas as pd
```

```python
def get_saved_results():
    with open("./final_results.json", mode ='r') as f:
        results = json.load(f)
    return results
results
```

```python
# parameters
x_as_len = 10 # used for smoothing
n_experiments = 10

def smooth(listX, old_x_as, new_x_as): 
    # from 10 points to 100 then old is the list of size 10 and new the list of size 100
    listX = make_interp_spline(old_x_as, listX, k=2)  # type: BSpline
    listX = listX(new_x_as)
    return listX
x_new = np.linspace(0, n_experiments-1, x_as_len)
x_old = np.linspace(0, n_experiments-1, 10)


results = get_saved_results()
for enemy in results.keys():
    meanAverage = np.array([])
    stdMean = np.array([])
    maxAverage = np.array([])
    stdMax = np.array([])
    for run in results[enemy]:
        current     = results[enemy][run]
        meanAverage = np.append(meanAverage, np.mean(current.get('means')))
        stdMean     = np.append(stdMean, np.std(current.get('means')))
        maxAverage  = np.append(maxAverage, np.mean(current.get('maximums')))
        stdMax      = np.append(stdMax, np.std(current.get('maximums')))

#     if len(meanAverage) < 10:
#         break

    meanAverage = smooth(meanAverage,x_old,x_new)
    maxAverage = smooth(maxAverage,x_old,x_new)
    stdMean = smooth(stdMean,x_old,x_new)
    stdMax = smooth(stdMax,x_old,x_new)
    print(meanAverage)
    plt.figure(figsize=(12, 6), dpi=360)
    plt.tight_layout()    
    plt.plot(x_new, meanAverage, label='meanAverage', color='#1B2ACC')
    plt.fill_between(np.linspace(0, n_experiments-1, x_as_len), 
                     meanAverage-stdMean, meanAverage+stdMean, 
                     alpha=.2, edgecolor='#1B2ACC', facecolor='#089FFF')

    plt.plot(x_new, maxAverage, label='maxAverage', color='#CC4F1B')
    plt.fill_between(np.linspace(0, n_experiments-1, x_as_len),
                     maxAverage-stdMax, maxAverage+stdMax,
                     alpha=.2, edgecolor='#CC4F1B', facecolor='#FF9848')
    plt.legend()
    plt.show()

```
